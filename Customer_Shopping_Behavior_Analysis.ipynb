### Exploratory Data Analysis (EDA)

To answer the business question and support the overall goal of improving sales, customer satisfaction, and long-term loyalty, the first step is understanding the data itself. This section focuses on exploring the customer behavior dataset to identify early signals, patterns, and potential drivers behind customer decisions.

In this part of the project, we will:

* Review the dataset structure, shape, and data types
* Identify missing values and inconsistencies that may affect analysis
* Run descriptive statistics to understand spending behavior and product trends
* Explore customer demographics, sales channels (online vs. offline), purchase frequency, and product category performance
* Look for relationships between variables such as discounts, seasons, reviews, and payment choices

The outcome of this analysis will help shape feature selection, guide modeling decisions, and build a clear direction for marketing and product strategy recommendations later in the project.

```python
# Loading the dataset using pandas

import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
# checking first 5 rows
df.head()
```

```python
# checking structure
df.info()
```

```python
# checking summary statistics
df.describe(include='all')
```

```python
# checking if missing data or null values are present in the dataset

df.isnull().sum()
```

```python
# imputing missing values in Review Rating column with the median rating of the product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# creating a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# creating new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

## Connecting Python script to PostgreSQL
### Why We Connected Python to PostgreSQL

To support deeper analysis and simulate real business transactions, we connected our Python environment to a PostgreSQL database. Python allowed us to clean and prepare the raw dataset, while PostgreSQL gave us a structured space to organize the data, create relational tables, and run targeted queries. This setup reflects how retail businesses manage data in real operations, where transactions, customer profiles, and product details are stored in databases rather than flat files. By using PostgreSQL, we can ask more realistic business questions, compare customer segments, and extract insights related to loyalty, purchase drivers, and shopping behavior.

### Data Analysis using SQL (Business Transactions)

We performed structured analysis in PostgreSQL to answer key business questions, such as identifying high-value customer groups, tracking repeat purchases, comparing performance across product categories, and understanding how online vs. offline channels influence revenue and frequency.

```python
!pip install psycopg2-binary sqlalchemy
```

```python
from sqlalchemy import create_engine
from getpass import getpass

username = input("Enter PostgreSQL username: ")
password = getpass("Enter PostgreSQL password: ")
host = "localhost"
port = "5432"
database = "Customer Behavior"

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# loading DataFrame into PostgreSQL
table_name = "customer_shopping_behavior"   # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
